```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline

np.random.seed(seed=1) #for reproducibility 
```

###Note that the code below generates 100 random values. Although the code provided in the sample notebook produces 100 values as well, it produces only approximately 70 unique values. Also worth noting is that the method below does not cluster values at the low end of the 10-2000 range.

```python
distSizes = np.random.choice((range(10,2000)),100, replace=False)
print sorted(distSizes)
```

##Regarding below: can we please have some instruction on how to create/ use classes in class?

```python
#I really really really need to learn how to create/ use classes.
mymean = 150

dists = {}
means = {}
df = mymean

figmain = plt.figure(figsize = (10,10), facecolor = "w")
figmain.subplots_adjust(wspace = 0.4, hspace = 0.5)

###Chi Square###
dists['chisq'] = {}

for eachsize in distSizes:
    dists['chisq'][eachsize] = np.random.chisquare(df, size = eachsize)

means['chisq'] = {}

axchisq_mu_n = figmain.add_subplot(321)
for dist in dists['chisq'].iterkeys():
    means['chisq'][dist] = dists['chisq'][dist].mean()
    
    axchisq_mu_n.plot(dist, means['chisq'][dist], '.', c='k')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.grid(True)
    axchisq_mu_n.grid(True)
    axchisq_mu_n.set_title('Chi Squared')
axchisq_mu_n.plot([min(distSizes), max(distSizes)],[df,df], 'r')

###Gaussian###
dists['normal'] = {}
mu = mymean
sigma = mymean  

for eachsize in distSizes:
    dists['normal'][eachsize] = np.random.normal(mu,sigma,eachsize)
    
means['normal'] = {}

axnorm = figmain.add_subplot(322)
for dist in dists['normal'].iterkeys():
    means['normal'][dist] = dists['normal'][dist].mean()
    
    axnorm.plot(dist, means['normal'][dist], '.', c='k')
    axnorm.set_xlabel('sample size')
    axnorm.set_ylabel('sample mean')
    axnorm.xaxis.grid(True)
    axnorm.yaxis.grid(True)
    axnorm.set_title("Normal")
axnorm.plot([min(distSizes), max(distSizes)],[mymean,mymean], 'r')

###Poisson###
dists['poisson'] = {}
lam = mymean

for eachsize in distSizes:
    dists['poisson'][eachsize] = np.random.poisson(lam, eachsize)
    
means['poisson'] = {}

axpois = figmain.add_subplot(323)
for dist in dists['poisson'].iterkeys():
    means['poisson'][dist] = dists['poisson'][dist].mean()
    
    axpois.plot(dist, means['poisson'][dist], '.', c='k')
    axpois.set_xlabel('sample size')
    axpois.set_ylabel('sample mean')
    axpois.xaxis.grid(True)
    axpois.yaxis.grid(True)
    axpois.set_title('Poisson')
axpois.plot([min(distSizes), max(distSizes)], [mymean, mymean], 'r') 

###Binomial###
dists['binomial'] = {}
bi_p = 0.5
bi_n = mymean/bi_p

for eachsize in distSizes:
    dists['binomial'][eachsize] = np.random.binomial(bi_n, bi_p, eachsize)

means['binomial'] = {}

axbin = figmain.add_subplot(324)
for dist in dists['binomial'].iterkeys():
    means['binomial'][dist] = dists['binomial'][dist].mean()
    
    axbin.plot(dist, means['binomial'][dist], '.', c='k')
    axbin.set_xlabel('sample size')
    axbin.set_ylabel('sample mean')
    axbin.xaxis.grid(True)
    axbin.yaxis.grid(True)
    axbin.set_title('Binomial')
axbin.plot([min(distSizes), max(distSizes)], [mymean, mymean], 'r') 

###Wald###
dists['wald'] = {}
mu = mymean
w_scale = 3

for eachsize in distSizes:
    dists['wald'][eachsize] = np.random.wald(mu, w_scale, eachsize)
    
means['wald'] = {}

axwald = figmain.add_subplot(325)
for dist in dists['wald'].iterkeys():
    means['wald'][dist] = dists['wald'][dist].mean()
    
    axwald.plot(dist, means['wald'][dist], '.', c='k')
    axwald.set_xlabel('sample size')
    axwald.set_ylabel('sample mean')
    axwald.xaxis.grid(True)
    axwald.yaxis.grid(True)
    axwald.set_title('Wald')
axwald.plot([min(distSizes), max(distSizes)],[df,df], 'r')

###Logistic###
dists['logistic'] = {}

loc = mymean
scale = mymean

for eachsize in distSizes: 
    dists['logistic'][eachsize] = np.random.logistic(loc, scale, eachsize)
    
means['logistic'] = {}

axlog = figmain.add_subplot(326)
for dist in dists['logistic'].iterkeys():
    means['logistic'][dist] = dists['logistic'][dist].mean()
    axlog.plot(dist, means['logistic'][dist], '.', c='k')
    axlog.set_xlabel('sample size')
    axlog.set_ylabel('sample mean')
    axlog.xaxis.grid(True)
    axlog.yaxis.grid(True)
    axlog.set_title('Logistic')
axlog.plot([min(distSizes), max(distSizes)], [loc, loc], 'r')

plt.tight_layout()
```

##As plotted above, the six distributions demonstrate that, generally, as sample size increases, its mean moves closer to the population mean. It is also evident that all distributions are somewhat bell-shaped.

##Optional: all means across all distributions in a histogram

```python
from scipy.stats import norm
import matplotlib.mlab as mlab

allmeans = []
for disttype in means: 
    for eachmean in means[disttype]:
        allmeans.append(means[disttype][eachmean])

#for histogram
plt.figure(figsize = (10,10))
n, bins, patches = plt.hist(allmeans, bins=50, normed=1, color='yellow')
plt.title('Distribution of means', fontsize=22)
plt.ylabel('Frequency', fontsize=18)
plt.xlabel('Mean', fontsize=18)

#for curve 
mu, sigma = norm.fit(allmeans)
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'k--')

plt.show()
```

###The above histogram demonstrates that across all distributions, the larger the sample size, the closer to the population mean it is. The dashed red line indicates that the distribution is highly gaussian.
